### **Load Data**

```python
import pandas as pd

# Load data set

def load_data(file_path):
    df = pd.read_csv(file_path)
    
    return df

file_path = '../data/flight_df.csv'
df = load_data(file_path)

df.head()
```

### **Exploratory Data Analysis**

```python
# Explore Dataset
def explore_data(df):
    
    num_rows = df.shape[0]
    num_cols = df.shape[1]
    data_types = df.info()
    
    print(f'Number of Rows: {num_rows}')
    print(f'Number of Columns: {num_cols}')
    print('--' * 30)
    print(f'Data Types: {data_types}')
    
explore_data(df)
```

```python
# EDA
def perform_eda(df):
    duplicates = df.duplicated().sum()
    missing_values = df.isna().sum()
    
    print(f'Number of duplicates: {duplicates}')
    print(f'Number of missing values: {missing_values}')
    
perform_eda(df)
```

```python
# Summary statistics

summary_stats = df.describe().T
print('Summary statistics:')
summary_stats
```

```python
# Check for anomalous values

def check_column_values(df):
    
    for column_name in df.columns:
        print(f'Unique values in {column_name} column:\n')
        unique_values = df[column_name].unique()
        print(unique_values)
        print('--' * 30)
        
check_column_values(df)
```

### **Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dark mode
plt.style.use('dark_background')

# Correlation Heatmap
def correlation_heatmap(df):
    fig, ax = plt.subplots(figsize=(15, 10))
    
    #define mask
    mask = np.triu(np.ones_like(df.corr(), dtype=bool))
    
    # Plot heatmap
    sns.heatmap(
        df.corr(),
        ax=ax,
        mask=mask,
        vmin=-1,
        vmax=1,
        cmap='BrBG',
        annot=True
    )
    
    ax.set_title('Triangle Correlation Heatmap')
    plt.show()
    
correlation_heatmap(df)
```

```python
# Ranked correlation heatmap
def ranked_heatmap(df, target_column):
    plt.figure(figsize=(5,10))
    
    # plot the heatmap
    heatmap = sns.heatmap(
        df.corr()[[target_column]].sort_values(by=target_column, ascending=False),
        vmin=-1,
        vmax=1,
        cmap='BrBG',
        annot=True
    )
    
    heatmap.set_title(f'Features Correlating with {target_column}', fontsize=16)
    plt.show()
    
ranked_heatmap(df, target_column='satisfied')
```

```python
# Create Histogram and boxplot
def visualize_data(df, column_name):
    fig, axes = plt.subplots(
        figsize = (10, 5),
        nrows=2,
        sharex=True
    )
    
    sns.set_style('darkgrid')
    
    # Histogram
    sns.histplot(
        df[column_name],
        ax=axes[0],
        bins='auto',
        kde = True
    )
    
    axes[0].tick_params(axis='x', rotation=45)
    
    # Boxplot
    sns.boxplot(
        data=df,
        x=column_name,
        ax=axes[1]
    )
    
    plt.tight_layout()
    plt.show()
    
visualize_data(df, column_name='Flight Distance')
    
```
